**SEABORN ASSIGNMENT:**

1. To create a scatter plot visualizing the relationship between two variables with synthetic data

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Generate synthetic data
np.random.seed(0)
x = np.random.rand(100)
y = 2 * x + np.random.normal(0, 0.1, 100)  # Adding some noise

# Create scatter plot
sns.scatterplot(x=x, y=y)
plt.title("Scatter Plot of Synthetic Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

2. To visualize the distribution of a numerical variable:

```python
# Generate random data
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot distribution
sns.histplot(data, kde=True)
plt.title("Distribution of Randomly Generated Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

3. To compare different categories based on numerical values:

```python
# Create data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 35, 20]

# Create a bar plot
sns.barplot(x=categories, y=values)
plt.title("Comparison of Categories by Value")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

4. To visualize the distribution of a numerical variable across different categories:

```python
# Generate synthetic data
data = np.random.normal(size=(100, 3)) + np.array([1, 2, 3])  # Different means for categories
categories = ['Category A'] * 100 + ['Category B'] * 100 + ['Category C'] * 100

# Create a box plot
sns.boxplot(x=categories, y=data.ravel())
plt.title("Distribution of Values Across Categories")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

5. To create a correlation heatmap for a synthetic dataset with correlated features:

```python
# Generate correlated data
data = np.random.multivariate_normal([0, 0, 0], [[1, 0.8, 0.5], [0.8, 1, 0.3], [0.5, 0.3, 1]], size=100)
df = pd.DataFrame(data, columns=["Feature 1", "Feature 2", "Feature 3"])

# Calculate the correlation matrix
corr_matrix = df.corr()

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

**PLOTLY ASSIGNMENT:**

1. 3D Scatter Plot

To visualize the distribution of data points in 3D space:

```python
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

fig = px.scatter_3d(df, x='X', y='Y', z='Z', title="3D Scatter Plot of Data Points")
fig.show()
```

2. Violin Plot for Student Grades

To display the distribution of scores across grade categories:

```python
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)

fig = px.violin(df, x="Grade", y="Score", box=True, points="all", title="Distribution of Scores by Grade")
fig.show()
```

3. Sales Heatmap

To visualize the variation in sales across different months and days:

```python
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

pivot_table = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

fig = px.imshow(pivot_table, title="Sales Variation Across Months and Days", labels={'color': 'Sales'})
fig.show()
```

4. 3D Surface Plot

To create a 3D surface plot for the function

```python
import plotly.graph_objects as go
import numpy as np

# Generate the data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create the surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title="3D Surface Plot of z = sin(sqrt(x^2 + y^2))", scene=dict(
                    xaxis_title='X',
                    yaxis_title='Y',
                    zaxis_title='Z'))
fig.show()
```

5. Bubble Chart for Country Data

To represent each country's population and GDP with bubble size proportional to population:

```python
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

fig = px.scatter(df, x='GDP', y='Population', size='Population', text='Country', title="Country Population vs GDP",
                 labels={'Population': 'Population (in millions)', 'GDP': 'GDP (in billions)'})
fig.show()
```

**BOKEH ASSIGNMENT:**

1. Sine Wave Plot

To display a sine wave:

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()  # Use this if you're running in a Jupyter notebook

# Generate sine wave data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='Y', width=600, height=400)
p.line(x, y, line_width=2, color='blue', legend_label='y = sin(x)')

show(p)
```

2. Scatter Plot with Random Values

To create a scatter plot using randomly generated values:

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import pandas as pd

output_notebook()

# Generate random data
np.random.seed(0)
n = 100
x = np.random.rand(n) * 10
y = np.random.rand(n) * 10
sizes = np.random.randint(10, 50, size=n)  # Marker sizes
colors = np.random.choice(['red', 'green', 'blue', 'orange'], size=n)  # Marker colors

# Create the plot
p = figure(title="Random Scatter Plot", width=600, height=400)
p.scatter(x, y, size=sizes, color=colors, alpha=0.6)

show(p)
```

3. Bar Chart for Fruit Counts

To create a bar chart representing counts of different fruits:

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

# Data for fruits
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create the plot
p = figure(x_range=fruits, title="Fruit Counts", width=600, height=400)
p.vbar(x=fruits, top=counts, width=0.5, color="green")

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = 'Count'
p.xaxis.axis_label = 'Fruits'

show(p)
```

4. Histogram

To create a histogram to visualize the distribution of random data:

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

# Generate histogram data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create the plot
p = figure(title="Histogram of Random Data", width=600, height=400)
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='blue', line_color='black', alpha=0.7)

# Customize the plot
p.xaxis.axis_label = 'Value'
p.yaxis.axis_label = 'Frequency'

show(p)
```

5. Heatmap

To create a heatmap using the provided dataset:

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource
from bokeh.palettes import Viridis256
import numpy as np

output_notebook()

# Generate heatmap data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the arrays
x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, value=z_flat))

# Create the plot
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', width=600, height=400)

# Create the heatmap
mapper = linear_cmap(field_name='value', palette=Viridis256, low=z_flat.min(), high=z_flat.max())
p.rect(x='x', y='y', width=0.1, height=0.1, source=source, line_color=None, fill_color=mapper)

# Add color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

show(p)
```
