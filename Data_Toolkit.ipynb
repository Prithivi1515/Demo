3. Explanation of Terms
* np.array, np.asarray, and np.asanyarray:

np.array: Creates a new array from a given object (list, tuple, etc.) and copies the data by default. It can create an array of any shape and type.

np.asarray: Similar to np.array, but it does not copy the data if the input is already an array. It is more memory efficient when you are sure that the input is an array.

np.asanyarray: Similar to np.asarray, but it allows the input to be a subclass of ndarray (like masked arrays). This means it returns the same subclass if possible.

* Deep Copy vs. Shallow Copy:

Deep Copy: Creates a new object that is a copy of the original object and recursively copies all objects found in the original. Changes made to the new object do not affect the original.

Shallow Copy: Creates a new object that is a copy of the original object but does not recursively copy the objects found in the original. Changes made to mutable objects in the shallow copy may affect the original object.

5. Creating a NumPy Array with Random Integers

```python
# Creating a 5x6 array with random integers between 1 and 10
random_integers = np.random.randint(1, 11, (5, 6))
print("\n5x6 array with random integers between 1 and 10:")
print(random_integers)

# a) Extract all even integers
even_integers = random_integers[random_integers % 2 == 0]
print("\nEven integers from the array:")
print(even_integers)

# b) Extract all odd integers
odd_integers = random_integers[random_integers % 2 != 0]
print("\nOdd integers from the array:")
print(odd_integers)
```

6. Creating a 3D NumPy Array of Random Integers

```python
# Creating a 3D array of shape (3, 3, 3) with random integers between 1 and 10
array_3d = np.random.randint(1, 11, (3, 3, 3))
print("\n3D array with random integers between 1 and 10:")
print(array_3d)

# a) Find the indices of the maximum values along each depth level (third axis)
max_indices = np.argmax(array_3d, axis=2)
print("\nIndices of maximum values along the third axis:")
print(max_indices)

# b) Element-wise multiplication of the array with itself
elementwise_multiplication = np.multiply(array_3d, array_3d)
print("\nElement-wise multiplication of the array with itself:")
print(elementwise_multiplication)
```

13. Scatter Plot of Two Random Float Arrays in Matplotlib

```python
import matplotlib.pyplot as plt

# Generate two random float arrays
x = np.random.rand(100)
y = np.random.rand(100)

# a) Create scatter plot
plt.scatter(x, y, color='red', marker='o', label='Data Points')

# b) Add horizontal line
plt.axhline(0.5, color='blue', linestyle='--', label='y = 0.5')

# c) Add vertical line
plt.axvline(0.5, color='green', linestyle=':', label='x = 0.5')

# d) Label axes and set title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Advanced Scatter Plot of Random Values')

# f) Display legend
plt.legend()

plt.show()
```

14. Time-Series Dataset with Temperature and Humidity


```python
import matplotlib.dates as mdates

# Create a sample time-series DataFrame
date_rng = pd.date_range(start='2024-01-01', end='2024-12-31', freq='D')
temperature = np.random.uniform(10, 35, len(date_rng))
humidity = np.random.uniform(30, 80, len(date_rng))

time_series_df = pd.DataFrame({'Date': date_rng, 'Temperature': temperature, 'Humidity': humidity})

# Plot Temperature and Humidity on the same plot with dual y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Temperature
ax1.plot(time_series_df['Date'], time_series_df['Temperature'], color='r', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color='r')
ax1.tick_params(axis='y', labelcolor='r')

# Plot Humidity on the second y-axis
ax2 = ax1.twinx()
ax2.plot(time_series_df['Date'], time_series_df['Humidity'], color='b', label='Humidity')
ax2.set_ylabel('Humidity', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# Set title and show plot
plt.title('Temperature and Humidity Over Time')
fig.tight_layout()
plt.show()
```

15. Histogram of Normal Distribution with PDF Overlay


```python
from scipy.stats import norm

# Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot histogram with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Plot PDF overlay
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Label axes and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()

plt.show()
```

17. Seaborn Scatter Plot with Quadrant Coloring


```python
import seaborn as sns

# Generate two random arrays for scatter plot
x_data = np.random.uniform(-1, 1, 100)
y_data = np.random.uniform(-1, 1, 100)

# Determine colors based on quadrant
colors = ['red' if (x > 0 and y > 0) else
          'blue' if (x > 0 and y < 0) else
          'green' if (x < 0 and y > 0) else
          'purple' for x, y in zip(x_data, y_data)]

# Create scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x_data, y=y_data, hue=colors, palette='deep', legend='full', s=50)

# Label and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Add legend
plt.legend(['Quadrant 1', 'Quadrant 2', 'Quadrant 3', 'Quadrant 4'], title='Quadrants')
plt.show()
```

18. Bokeh Line Plot of a Sine Wave Function


```python
from bokeh.plotting import figure, show, output_notebook
import math

# Prepare data
x_vals = np.linspace(0, 4 * math.pi, 100)
y_vals = np.sin(x_vals)

# Create plot
output_notebook()
p = figure(title="Sine Wave Function", x_axis_label="X", y_axis_label="Y")

# Add sine wave line and grid
p.line(x_vals, y_vals, line_width=2)
p.grid.grid_line_color = "gray"

show(p)
```

19. Bokeh Bar Chart with Categorical Data


```python
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure, show
import random

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [random.randint(1, 100) for _ in categories]
source = ColumnDataSource(data=dict(categories=categories, values=values, color=['blue', 'green', 'red', 'purple']))

# Create bar chart
output_notebook()
p = figure(x_range=categories, title="Random Categorical Bar Chart", toolbar_location=None, tools="")

p.vbar(x='categories', top='values', width=0.5, color='color', source=source)

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Labels
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"
show(p)
```
